```python
import pandas as pd 
import numpy  as np 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("olxcleaned.csv")
```

```python
lb = LabelEncoder()
df['Make_Encode'] = lb.fit_transform(df['Make'])
```

```python
lb = LabelEncoder()
feature = {}
for k in df.columns:
    if df[k].dtypes=='O':
        df[str(k)+'_Encode'] = lb.fit_transform(df[k])
        gt = {}
        ke = df[[k,str(k)+'_Encode']].drop_duplicates()[str(k)+'_Encode'].values
        val = df[[k,str(k)+'_Encode']].drop_duplicates()[k].values
        for i in range(len(ke)):
            gt[val[i]] = str(ke[i])
        feature[k] = gt
```

```python
feature
```

```python
import json
with open('encdata.json','w') as jsonfile:
    json.dump(feature,jsonfile,indent=4)
```

```python
df.drop(["Make","Model","Fuel","Registration city","Car documents","Assembly","Transmission","Condition"],axis = 1,inplace=True)
```

```python
df
```

```python
y = df['Price']
x = df.drop(['Price','Condition_Encode'],axis = 1)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
df.drop(['Price','Condition_Encode'],axis=1,inplace=True)
```
